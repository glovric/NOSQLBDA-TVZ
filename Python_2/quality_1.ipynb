# Distribution plotting

```python
def random_color():
    return (np.random.random(), np.random.random(), np.random.random())

features_units = [
    "microsiemens per centimeter",
    "standard unit",
    "standard unit",
    "microsiemens per centimeter",
    "microsiemens per centimeter",
    "milligrams per liter",
    "milligrams per liter",
    "milligrams per liter",
    "degrees celsius",
    "degrees celsius",
    "degrees celsius",
    ""
]

fig, axes = plt.subplots(4, 3, figsize=(12, 10))
axes = axes.flatten()

for idx, c in enumerate(df.columns):
    df[c].plot(kind='hist', ax=axes[idx], edgecolor="black", color=random_color(), bins=20)
    axes[idx].set_title(c)
    axes[idx].set_xlabel(features_units[idx])

plt.tight_layout()
plt.show()
```

- Specific conductance (Maximum): right skewed, irregular

- pH (Maximum): normal-like, centered around 0.9

- ph (Minimum): right skewed, irregular

- Specific conductance (Minimum): right skewed, irregular

- Specific conductance (Mean): normal-like, centered around 0.5

- Dissolved oxygen (Maximum): normal-like, centered around 0.85

- Dissolved oxygen (Mean): normal-like, centered around 0.55

- Dissolved oxygen (Minimum): normal-like, centered around 0.6

- Temperature (Mean): left-skewed

- Temperature (Minimum): left-skewed

- Temperature (Maximum): left-skewed

- Target: normal-like, centered around 0.65

# Target variable inspection

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 4))
axes = axes.flatten()

df["Target"].plot(kind='hist', edgecolor="black", color=random_color(), bins=30, ax=axes[0])
axes[0].set_title("Hist plot")

df["Target"].plot(kind='density', ax=axes[1], color=random_color())
axes[1].set_title("Density plot")

df["Target"].plot(kind='box', ax=axes[2])
axes[2].set_title("Box plot")

plt.tight_layout()
plt.show()
```

- given Hist and Density plots, we observe most values concentrated around 0.65

- from the Box plot we observe 1st, 2nd and 3rd quartiles being between 0.65 and 0.68, where 50% of data is contained

- below 0.6 and above 0.7 we have many outliers which are beyon the whiskers
